# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

### Steps of Hypothesis Testing

1. SET THE HYPOTESIS 
2. CHOOSE SIGNIFICANCE LEVEL
3. SAMPLE
4. COMPUTE THE STATISTIC 
5. GET p-VALUE
6. DECIDE

```python
# 1. SET THE HYPOTESIS 

##NOTE: mu -> is the average of the population 

# H0: mu the hourly workers is equal to USD 30/hr 
# H1: mu the hourly workers is different from USD 30/hr

## Important: two tailed test
```

```python
# 2. CHOOSE THE SIGNIFICANCE LEVEL

# significance level = 5% (which is 100-confidence level)

alpha = 0.05
```

```python
# 3. SAMPLE of the hourly workers

hourly_sample = salaries[salaries['Salary or Hourly'] == 'Hourly']
hourly_sample
```

```python
hourly_sample.shape
```

```python
# Mean of our sample (x_hat)
x_hat = hourly_sample['Hourly Rate'].mean() 
x_hat
```

```python
# 4. COMPUTE THE STATISTICS
# 5. COMPUTE THE P_VALUE

stat, p_value = st.ttest_1samp(hourly_sample['Hourly Rate'], 30)

print(stat)
print(p_value)
```

```python
#6. DECIDE

p_value < alpha
```

##### Considerations

1. We center our standard distribution on the mean of the our sample, which is the column Hourly Rate of the dataframe hourly_sample

2. Our statistic is 20 positive standard deviation times away from the mean of  my population

3. Therfore: we prefer to reject H0, 
- Since the standard deviation is very far away from the mean of my population
- Since the computed p_value is much smaller than the significant level set a priori, our alpha of 5%

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our hypothesis is greater than, then to reject, the test statistic must also be positive.

```python
# 1. SET THE HYPOTESIS 

##NOTE: mu -> is the average of the population 

# H0: mu the annual salaries this year are lower or equal to last year's mean of USD 86k
# H1: mu the annual salaries this year are higher that last year's mean of USD 86k

## Important: one tailed test
```

```python
# 2. CHOOSE THE SIGNIFICANCE LEVEL

# significance level = 5% (which is 100-confidence level)

alpha = 0.05
```

```python
# 3. SAMPLE of the police force

condition = (salaries['Department'] == 'POLICE') & (salaries['Salary or Hourly'] == 'Salary')
```

```python
police_sample = salaries[condition]
police_sample.head(10)
```

```python
police_sample.shape
```

```python
# Mean of our sample (x_hat)

x_hat = police_sample['Annual Salary'].mean() 
x_hat
```

```python
# 4. COMPUTE THE STATISTICS
# 5. COMPUTE THE P_VALUE

stat, p_value = st.ttest_1samp(police_sample['Annual Salary'], 86000.0, alternative='less')

print(stat)
print(p_value)
```

```python
# 6. DECISION

p_value < alpha
```

##### Considerations

1. We center our standard distribution on the mean of the our sample, which is the column Annual Salary of the dataframe police_sample

2. Our statistic is 3 positive standard deviation times away from the mean of my population

3. Therfore: we prefer to not reject H0, 
- Since the standard deviation is very close from the mean of my population
- Since the computed p_value is bigger than the significant level set a priori, our alpha of 5%

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers.

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale.
